```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data Loading
df = pd.read_csv("Customer_Behavior_Analysis.csv")
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Print the correlation matrix
print("\nCorrelation matrix:")
print(correlation_matrix)

# Data Exploration
print("Shape of the dataset:", df.shape)
print("\nColumns in the dataset:", df.columns)
print("\nSummary statistics of numerical columns:")
print(df.describe())
print("\nInformation about the dataset:")
print(df.info())

# Visualizations
# Let's visualize the distribution of Satisfaction Rating
plt.figure(figsize=(8, 6))
sns.countplot(x='Satisfaction Rating', data=df)
plt.title('Distribution of Satisfaction Rating')
plt.xlabel('Satisfaction Rating')
plt.ylabel('Count')
plt.show()

# Let's visualize the relationship between Energy Consumed and Charging Time
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Charging Time (mins)', y='Energy Consumed (kWh)', data=df)
plt.title('Energy Consumed vs Charging Time')
plt.xlabel('Charging Time (mins)')
plt.ylabel('Energy Consumed (kWh)')
plt.show()

# Statistical Analysis
# Calculate the average Energy Consumed per Charging Station
avg_energy_per_station = df.groupby('Charging Station ID')['Energy Consumed (kWh)'].mean()
print("\nAverage Energy Consumed per Charging Station:")
print(avg_energy_per_station)

# Calculate the correlation matrix
correlation_matrix = df.corr()
print("\nCorrelation matrix:")
print(correlation_matrix)
```
